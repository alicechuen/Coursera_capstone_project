This information can be used if an alternate company wants to bid against SpaceX for a rocket launch.

This dataset includes a record for each payload carried during a SpaceX mission into outer space.

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
mycursor.execute("SELECT DISTINCT(Launch_Site) FROM spacex")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
mycursor.execute("SELECT * FROM spacex  WHERE Launch_Site LIKE 'CCA%' LIMIT 5")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
mycursor.execute("SELECT SUM(PAYLOAD_MASS__KG_) FROM spacex  WHERE CUSTOMER = 'NASA (CRS)'")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
mycursor.execute("SELECT AVG(PAYLOAD_MASS__KG_) FROM spacex  WHERE Booster_Version LIKE 'F9 v1.1%'")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
mycursor.execute("SELECT min(DATE) FROM spacex  WHERE Landing_Outcome = 'Success'")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
mycursor.execute("SELECT Booster_Version FROM spacex  WHERE Landing_Outcome = 'Success' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
mycursor.execute("SELECT LANDING_OUTCOME, COUNT(*) FROM spacex GROUP BY LANDING_OUTCOME ")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
mycursor.execute("SELECT Booster_Version FROM SPACEX WHERE (SELECT MAX(PAYLOAD_MASS__KG_) FROM spacex)  ")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
mycursor.execute("SELECT Booster_Version, LAUNCH_SITE FROM SPACEX WHERE LANDING_OUTCOME = 'Failure (drone ship)' AND DATE BETWEEN '2015-01-01' AND '2015-12-31'")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
mycursor.execute("SELECT *, COUNT(*) C FROM SPACEX WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY LANDING_OUTCOME ORDER BY C DESC ")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```
